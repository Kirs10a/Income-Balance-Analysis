```python
# Import pandas
import pandas as pd
```

```python
# Create a DataFrame with balance.txt
# Read in information from balance.txt
balance_df = pd.read_csv('balance.txt', sep=' ')
```

Average income, based on ethnicity

```python

# Group the average income on ethnicity
ethnicity_based_income = balance_df.groupby('Ethnicity')['Income'].mean()

# Sort the ethnicity groups and income from highest to lowest
ethnicity_based_income_sorted = ethnicity_based_income.sort_values(ascending=False)

# Print output in descending order
print(f"Average income by ethnicity, sorted from highest to lowest: "
      f" \n{ethnicity_based_income_sorted}")
```

```python
# On average, do married or single people have a higher balance?
# Calculate the average balance for each group 

average_balance_by_marriage = balance_df.groupby('Married')['Balance'].mean()

# Print the average balance for married and single individuals
print("Average balance by marital status:")
print(average_balance_by_marriage)

# Compare the average balances
average_balance_married = average_balance_by_marriage.loc['Yes']
average_balance_single = average_balance_by_marriage.loc['No']

if average_balance_married > average_balance_single:
    print("\nOn average, married people have a higher balance.")
elif average_balance_married < average_balance_single:
    print("\nOn average, single people have a higher balance.")
else:
    print("\nOn average, both married and single people have the same balance.")
```

Highest Income

```python
# What is the lowest income in our dataset?

# Find the highest income 
highest_income = balance_df['Income'].max()
# Print the highest income, rounded to 2 decimals 
print(f"The highest income: {highest_income:.2f}\n")

# Print the details of the person with the highest income by finding
# locaation of the highest income 
row_highest_income = balance_df.loc[balance_df['Income'].idxmax()]
print(f"Details of the individual with the highest income:\n{row_highest_income}")
```

Lowest Income 

```python
# Find the lowest income 
lowest_income = balance_df['Income'].min()

# Print the lowest income, rounded to 2 decimals 
print(f"The lowest income: {lowest_income:.2f}\n")

# Print the details of the person with the lowest income by finding
# locaation of the lowest income 
row_lowest_income = balance_df.loc[balance_df['Income'].idxmin()]
print(f"Details of the individual with the lowest income:\n{row_lowest_income}")
```

Number of cards recorded in data set.

```python
# Calculate the sum of the Cards column
number_cards = balance_df['Cards'].sum()

# Print total number of cards
print(f"Total number of cards: {number_cards}")
```

Number of females vs males in dataset

```python
# Count the number of females and males
num_females = balance_df[balance_df['Gender'] == 'Female'].count()['Gender']
num_males = balance_df[balance_df['Gender'] == 'Male'].count()['Gender']

# Print the results
print(f"Number of females: {num_females}")
print(f"Number of males: {num_males}")
```
